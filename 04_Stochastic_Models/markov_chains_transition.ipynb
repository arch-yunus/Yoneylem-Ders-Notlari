# Markov Zincirleri ve Durum Geçişleri 🎲

Bu notebook, bir sistemin farklı durumlar arasındaki geçiş olasılıklarını ve kararlı durum (steady-state) analizini içerir.

## Temel Kavramlar:
- **Durum Uzayı:** Sistemin alabileceği tüm olası durumlar.
- **Geçiş Matrisi (P):** $P_{ij}$, sistemin $i$ durumundan $j$ durumuna geçme olasılığı.
- **Kararlı Durum:** Uzun vadede sistemin her bir durumda bulunma olasılığının sabitlenmesi.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx

# Geçiş Matrisi Tanımlama
# Durumlar: 0: Güneşli, 1: Bulutlu, 2: Yağmurlu
P = np.array([
    [0.7, 0.2, 0.1],
    [0.3, 0.4, 0.3],
    [0.2, 0.3, 0.5]
])

states = ["Güneşli", "Bulutlu", "Yağmurlu"]
df_transition = pd.DataFrame(P, index=states, columns=states)
print("Geçiş Matrisi:")
df_transition
```

## Kararlı Durum Hesaplama
Uzun vadeli olasılıkları $(\pi P = \pi)$ denklemini çözerek bulalım.

```python
# Özvektör yöntemi ile kararlı durum
vals, vecs = np.linalg.eig(P.T)
stationary = vecs[:, np.isclose(vals, 1)].real
stationary = stationary / stationary.sum()

print("Kararlı Durum Olasılıkları:")
for i, state in enumerate(states):
    print(f"{state}: {stationary[i][0]:.4f}")
```

## Durum Geçiş Diyagramı Görselleştirmesi

```python
G = nx.MultiDiGraph()
for i in range(len(states)):
    for j in range(len(states)):
        if P[i, j] > 0:
            G.add_edge(states[i], states[j], weight=P[i, j], label=f"{P[i, j]:.2f}")

pos = nx.spring_layout(G)
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="orange", font_size=10, arrowsize=20)
edge_labels = nx.get_edge_attributes(G, 'label')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
plt.title("Markov Zinciri Durum Geçiş Diyagramı")
plt.show()
```
