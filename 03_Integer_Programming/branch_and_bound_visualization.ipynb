# Dal ve Sınır (Branch and Bound) Algoritması 🌳

Bu notebook, Tamsayılı Programlama problemlerini çözmek için kullanılan **Dal ve Sınır** algoritmasının mantığını görselleştirir.

## Algoritma Mantığı:
1. **Gevşetme (Relaxation):** Tamsayı kısıtları kaldırılarak problem bir Doğrusal Programlama problemi gibi çözülür.
2. **Dallanma (Branching):** Eğer sonuç tamsayı değilse, değişken tamsayı olacak şekilde iki yeni alt probleme ayrılır.
3. **Sınırlandırma (Bounding):** Alt çözümler, şimdiye kadar bulunan en iyi tamsayı çözümle karşılaştırılır.

```python
import networkx as nx
import matplotlib.pyplot as plt

def draw_bb_tree():
    G = nx.Graph()
    
    # Düğümler (Adımlar)
    nodes = {
        "P0": "Ana Problem (Z=15.5)",
        "P1": "x1 <= 2 (Z=14.2)",
        "P2": "x1 >= 3 (Z=13.8)",
        "P1.1": "x2 <= 1 (Z=12.0) [TAM]",
        "P1.2": "x2 >= 2 (Uygunsuz)"
    }
    
    edges = [("P0", "P1"), ("P0", "P2"), ("P1", "P1.1"), ("P1", "P1.2")]
    
    pos = {
        "P0": (0, 0), 
        "P1": (-1, -1), "P2": (1, -1), 
        "P1.1": (-1.5, -2), "P1.2": (-0.5, -2)
    }
    
    plt.figure(figsize=(10, 6))
    nx.draw(G, pos, with_labels=False, node_size=3000, node_color="skyblue", font_size=10)
    
    for node, label in nodes.items():
        G.add_node(node)
        plt.text(pos[node][0], pos[node][1], label, 
                 fontsize=9, ha='center', va='center', 
                 bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))
    
    G.add_edges_from(edges)
    nx.draw_networkx_edges(G, pos, width=2, alpha=0.6, edge_color="gray")
    
    plt.title("Dal ve Sınır (Branch and Bound) Karar Ağacı Görselleştirmesi")
    plt.axis('off')
    plt.show()

draw_bb_tree()
```

--- 
Bu görselleştirme, algoritmanın tamsayı sonuca ulaşana kadar karar uzayını nasıl budadığını (pruning) temsil eder.
