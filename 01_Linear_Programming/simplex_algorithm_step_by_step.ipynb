# Simpleks Algoritması ile Adım Adım Optimizasyon 🚀

Bu notebook, Doğrusal Programlama problemlerini çözmek için kullanılan klasik **Simpleks Algoritmasını** hem teorik hem de uygulamalı (Python - SciPy) olarak açıklar.

```python
import numpy as np
from scipy.optimize import linprog
import pandas as pd

print("Kütüphaneler yüklendi.")
```

## 1. Örnek Problem Tanımı

Aşağıdaki maksimizasyon problemini ele alalım:

**Amaç Fonksiyonu:**
$$Z_{max} = 3x_1 + 5x_2$$

**Kısıtlar:**
1. $x_1 \leq 4$
2. $2x_2 \leq 12$
3. $3x_1 + 2x_2 \leq 18$
4. $x_1, x_2 \geq 0$

```python
# Katsayılar (linprog minimizasyon yaptığı için maksimizasyon katsayılarını negatif alıyoruz)
obj = [-3, -5]

# Kısıtlar sol taraf matrisi (Ax <= b)
lhs_ineq = [[ 1,  0],  # 1x1 + 0x2 <= 4
            [ 0,  2],  # 0x1 + 2x2 <= 12
            [ 3,  2]]  # 3x1 + 2x2 <= 18

# Kısıtlar sağ taraf vektörü (b)
rhs_ineq = [4, 
            12, 
            18]

# Değişken sınırları
bnd = [(0, float("inf")),  # x1 >= 0
       (0, float("inf"))]  # x2 >= 0

# Çözüm
opt = linprog(c=obj, A_ub=lhs_ineq, b_ub=rhs_ineq, bounds=bnd, method="highs")

print(f"Optimum Z Değeri: {-opt.fun}")
print(f"Karar Değişkenleri [x1, x2]: {opt.x}")
```

## 2. Sonuçların Analizi

Algoritma, kısıtların kesişim noktalarını (köşe noktalarını) tarayarak en yüksek karı getiren koordinatları bulmuştur.

> **İpucu:** Gerçek dünyada `x1` ve `x2` örneğin üretilen ürün miktarlarını, `Z` ise toplam karı temsil eder.
