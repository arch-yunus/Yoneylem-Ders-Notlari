# Duyarlılık Analizi (Sensitivity Analysis) 📊

Bu notebook, bir Doğrusal Programlama modelinin parametrelerindeki değişimlerin optimum çözümü nasıl etkilediğini görselleştirir. 

## Temel Kavramlar:
- **Gölge Fiyat (Shadow Price):** Bir kısıtın sağ taraf değeri ($b_i$) 1 birim arttığında amaç fonksiyonunda meydana gelen değişim.
- **Değişim Aralıkları:** Optimum çözümün değişmeden kaldığı katsayı aralıkları.

```python
import pulp
import pandas as pd
import matplotlib.pyplot as plt

# Basit bir üretim modeli kuralım
prob = pulp.LpProblem("Üretim_Optimizasyonu", pulp.LpMaximize)

# Karar değişkenleri
x1 = pulp.LpVariable("Ürün_A", 0)
x2 = pulp.LpVariable("Ürün_B", 0)

# Amaç Fonksiyonu
prob += 40 * x1 + 30 * x2, "Kar"

# Kısıtlar
prob += 2 * x1 + 1 * x2 <= 20, "İşçilik_Saati"
prob += 1 * x1 + 1 * x2 <= 15, "Hammadde"
prob += 0.5 * x1 + 1 * x2 <= 12, "Makine_Saati"

# Çözüm
prob.solve(pulp.PULP_CBC_CMD(msg=0))

print(f"Durum: {pulp.LpStatus[prob.status]}")
print(f"Optimum Kar: {pulp.value(prob.objective)}")
```

## Gölge Fiyat Analizi
Kısıtların gıda fiyatlarını (shadow prices) inceleyelim.

```python
o = [{'Kısıt': name, 'Gölge_Fiyat': c.pi, 'Gevşeklik': c.slack} 
     for name, c in prob.constraints.items()]
df_sensitivity = pd.DataFrame(o)
df_sensitivity
```

## Görselleştirme
Kısıtların hassasiyetini grafik üzerinde görelim.

```python
df_sensitivity.plot(kind='bar', x='Kısıt', y='Gölge_Fiyat', color='purple')
plt.title("Kısıtların Gölge Fiyat Analizi")
plt.ylabel("Gölge Fiyat (Pi)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
